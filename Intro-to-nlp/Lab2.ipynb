```python
import nltk
from nltk.corpus import brown
labels = ['i', 'he', 'she', 'we', 'they']
```

```python
brown_freq = nltk.FreqDist([w for w in brown.words() if w.lower() in labels])
```

```python
brown_freq.tabulate()
```

```python
brown_freq = nltk.FreqDist([w.lower() for w in brown.words() if w.lower() in labels])
```

```python
brown_freq.tabulate()
```

```python
brown_freq['I'] = brown_freq.pop('i')
```

```python
brown_freq.pop('i')
```

```python
brown_freq.tabulate()
```

```python
%matplotlib inline
brown_freq.plot()
```

```python
cond = nltk.ConditionalFreqDist([(genre, word.lower()) 
                                 for genre in ['news','fiction']
                                 for word in brown.words(categories=genre) 
                                 if word.lower() in labels])
```

```python
cond.tabulate()
```

```python
%matplotlib inline
cond.plot()
```
